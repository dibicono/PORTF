```python
import pandas as pd

df = pd.read_csv("worldcup2014.csv")
df.head(10)
```

```python
# se calculan la cantidad de puestos por continentes y paises
ContinentesUnicos = len(df["continent"].unique())
PaisesUnicos = len(df["team"].unique())
print(f"Son: {ContinentesUnicos} continentes y {PaisesUnicos} paises unicos en total")

#Contar cuantas los paises que son pertenecientes a cada contimnente
PPC = df["continent"].value_counts()
df_ppc = PPC.reset_index()
df_ppc.columns = ['continent', 'paises']
df_ppc
```

```python
#probabilidad de elegir un pais
p_pais = 1/32
print(f"probabilidad de elegir un pais = {p_pais}")
p_cont = 1/5
print(f"probabilidad de elegir un continente = {p_cont}")
p_europa = 13/32
p_suramerica = 6/32
p_africa = 5/32
p_norteamerica = 4/32
p_asia = 4/32

data = {
    'continent': ['europe', 'southamerica', 'africa', 'northamerica', 'asia'],
    'probabilidad': [13/32, 6/32, 5/32, 4/32, 4/32]
}

ProbSerDeUnPais = pd.DataFrame(data)
ProbSerDeUnPais


#print(f"probabilidad de que sea europa = {p_europa}\nprobabilidad de que sea suramerica = {p_suramerica}\nprobabilidad de que sea africa = {p_africa}\nprobabilidad de que sea norteamerica = {p_norteamerica}\nprobabilidad de que sea asia = {p_asia}")
```

a. pertenecer a África

```python
SeaAfrica = p_pais + p_africa - (p_africa * p_pais)
SeaAfrica
```

b. pertenecer a Asia

```python
SeaAsia = p_pais + p_asia - (p_asia * p_pais)
SeaAsia
```

c. pertenecer a Europa

```python
SeaEuropa = p_pais + p_europa - (p_europa * p_pais)
SeaEuropa
```

d. pertenecer a Norteamérica

```python
SeaNorteamerica = p_pais + p_norteamerica - (p_norteamerica * p_pais)
SeaNorteamerica
```

e. pertenecer a Sudamérica

```python
SeaSudamerica = p_pais + p_suramerica - (p_suramerica * p_pais)
SeaSudamerica
```

f. clasificar a segunda ronda

```python
#se suma quienes pasaron
TeamsSegunda = df["clasificado"].sum()
p_SegundaRonda = TeamsSegunda / PaisesUnicos
print(f"Probabilidad de pasar a la segunda ronda = {p_SegundaRonda}")
```

g. ganar al menos un partido

```python
df["Proba_Ganar"] = df["juegos_ganados"] / df["cantidad_juegos"]
df[['continent', 'team', "Proba_Ganar", "juegos_ganados", "cantidad_juegos"]]
```

```python
dataPaises = {"continent": ['europe', 'southamerica', 'africa', 'northamerica', 'asia'], 'Probabilidad_ser_continente': [p_europa, p_suramerica, p_africa, p_norteamerica, p_asia]}
ProbSerDeUncontinente = pd.DataFrame(dataPaises)
ProbSerDeUncontinente
```

```python
ProbSerDeUncontinente['Ser_Continente_y_Clasificar'] = ProbSerDeUncontinente['Probabilidad_ser_continente'] * p_SegundaRonda
print(f"Probabilidad de pasar a la segunda ronda = {p_SegundaRonda}")
ProbSerDeUncontinente[['continent', 'Ser_Continente_y_Clasificar']]
```

2. A partir de las probabilidades anteriores, y sin realizar conteos, calcular las
siguientes probabilidades:

a. pertenecer a Europa, si se sabe que clasificó
b. pertenecer a Sudamérica, si se sabe que clasificó


```python
df_unidas["SerContinente_y_Clasifico"] = (df_unidas['Probabilidad_ser_continente'] * (p_SegundaRonda * df_unidas['Probabilidad_ser_continente']))/p_SegundaRonda
df_unidas[['Probabilidad_ser_continente', "SerContinente_y_Clasifico", "team", 'continent']]
```

c. pertenecer a África, si se sabe que ganó al menos un partido


```python
df_unidas["SerContinente_y_gano1P"] = (df_unidas['Probabilidad_ser_continente'] * (df_unidas["Proba_Ganar"] * df_unidas['Probabilidad_ser_continente']))/df_unidas["Proba_Ganar"]
df_unidas[['Probabilidad_ser_continente', "SerContinente_y_gano1P", "team", 'continent']]
```

d. haber ganado al menos un partido, si se sabe que clasificó

```python
df_unidas["Ganar1P_y_Clasifico"] = (df_unidas["Proba_Ganar"] * (p_SegundaRonda * df_unidas["Proba_Ganar"]))/p_SegundaRonda
df_unidas[['Probabilidad_ser_continente', "Ganar1P_y_Clasifico", "team"]]
```

e. haber ganado al menos un partido, si se sabe que pertenece a Norteamérica

```python
df_unidas["Norteamerica_y_gano1P"] = (0.12500 * (df_unidas["Proba_Ganar"] * 0.12500))/df_unidas["Proba_Ganar"]
df_unidas[df_unidas["continent"] == "northamerica"][['Probabilidad_ser_continente', "Norteamerica_y_gano1P", "team", "continent"]]
```

f. NO haber clasificado, si se sabe que pertenece a África

```python
df_unidas["No_clasifico_y_Africa"] = ((1-p_SegundaRonda) * (0.15625 * (1-p_SegundaRonda)))/0.15625
df_unidas[df_unidas["continent"] == "africa"][['Probabilidad_ser_continente', "No_clasifico_y_Africa", "team", "continent"]]
```
