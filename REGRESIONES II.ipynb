5. Si de este dataset se escoge un paciente al azar, ¿cuál es la probabilidad de que sea hombre, si se sabe que pertenece al mayor cuartil de la variable correspondiente al índice de masa corporal?

```python
#Probabilidad de un datos estar en quartil para BMI
qmin = dfObs["BMI"].quantile(0.75)
cDatos = len(dfObs[dfObs["BMI"] >= qmin])
cTotal = len(dfObs["BMI"])
P_QBMI = cDatos / cTotal
print(f"Probabilidad de pertenecer al cuarto quartil: {P_QBMI*100:.2f}%")

#Probabilidad de que un paciente sea hombre
cHombre = len(dfObs[dfObs["sex"] == "M"])
P_QHOM = cHombre / cTotal
print(f"Probabilidad de pertenecer al genero masculino: {P_QHOM*100:.2f}%")

#Calculo de probabilidad dependiente
P_A_B = (cDatos-1) / (cTotal - 1)
P_dependiente = P_A_B * P_QBMI / P_QHOM
print(f"Probabilidad de pertenecer al genero masculino despues de haber estado en el quartil mayor: {P_dependiente*100:.2f}%")
```

6. ¿Es razonable afirmar que, para una persona cualquiera (no necesariamente dentro del dataset), su ritmo cardiaco promedio es 75? Explica y justifica.

```python
import scipy.stats as stats
#calculo datos descriptivos
promHR = dfObs["heartRate"].mean() #promedio
desvhr = dfObs["heartRate"].std() #desviacion

#creacion de datos aleatorios
np.random.seed(43)
datos = np.random.normal(loc=promHR, scale=2, size=1000)

#parametros
H1 = 75 #hipotesis
alpha = 0.05
n = len(dfObs["heartRate"]) #cantidad de datos
promHR_D = datos.mean()
desvhr_D = datos.std()

#t_student
t = (promHR - promHR_D) / (desvhr / np.sqrt(n))
df_a = n-1
p_valor =  2* (1-stats.t.cdf(abs(t), df_a))
print(f"Valor de p: {p_valor:.4f}, > {p_valor*100:.2f}%\nValor t: {t:.4f}")

#graficar
margen = 3
x_min = t - margen
x_max = t + margen
x2 = np.linspace (x_min, x_max, 300)
y2 = stats.t.pdf(x2, df_a)

#grafico
plt.figure(figsize=(6,3))
plt.plot(x2, y2, label="Distribución")
t_critico = stats.t.ppf(1 - alpha / 2, df_a)
plt.fill_between(x2, 0, y2, where=(x2 > t_critico), color="blue", alpha=0.5)
plt.fill_between(x2, 0, y2, where=(x2 < -t_critico), color="blue", alpha=0.5)

#linea de observacion h1
plt.axvline(t, color="green", linestyle="--", label=f"H1: {H1}")
plt.title('Distribución t y prueba de hipótesis (bilateral)', fontsize=14)
plt.legend(fontsize=9)
plt.xlabel('Valores t', fontsize=10)
plt.ylabel('Probabilidad', fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.xlim(-4, 4)

plt.show()
```

SI, La probabilidad de este caso en que el ritmo cardiaco se de 75 como hipotesis H1, es muy alta, es de 97.84% estimada, osea para cualquier persona que se escoja, la probabilidad de que su ritmo cardiaco sea de 75 es de 97% lo que hace que la hipotesis H1 sea totalmente posible.

8. Construye un modelo de regresión lineal que permita relacionar 6 variables del dataset con el índice de masa corporal. (Debes incluir al menos dos variables cualitativas). Evalúa tu modelo y explica.

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

#convertir booleanos a numeros 1 y 0
dfObs["M"] = dfObs["M"].astype("int") 
dfObs["F"] = dfObs["F"].astype("int") 

#eliminar los nan
dfObsRL = dfObs.dropna(subset=["totChol", "M", "F", "cigsPerDay", "sysBP", "heartRate", "BMI"])

#declarar las variables dependiente e independientes
x = dfObsRL[["totChol", "M", "F", "cigsPerDay", "sysBP", "heartRate"]].apply(pd.to_numeric, errors="coerce")
y = pd.to_numeric(dfObsRL["BMI"], errors="coerce")

#se crea modelo
X = sm.add_constant(x)
model = sm.OLS(y, X, missing="drop").fit()
print(model.summary())

# crear datos para linea ajustada de totChol
x_chol = np.linspace(x["totChol"].min(), x["totChol"].max(), 955)
X_line = pd.DataFrame({
    "totChol": x_chol,
    "M": x["M"].mean(),
    "F": x["F"].mean(),
    "cigsPerDay": x["cigsPerDay"].mean(),
    "sysBP": x["sysBP"].mean(),
    "heartRate": x["heartRate"].mean()
})

#modelo grafico
model = LinearRegression()
model.fit(x, y)
y_prediccion = model.predict(X_line)

#GRAFICA
plt.figure(figsize=(5,3))
plt.scatter(x["totChol"], y, label="Regresioin lineal",  s=2, alpha=0.5)
plt.plot(x_chol, y_prediccion, color="orange", label="Prediccion")
plt.xlabel("Colesterol total", fontsize=9)
plt.ylabel("IMC", fontsize=9)
plt.title("Regresión lineal múltiple", fontsize=14)
plt.legend(fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

El modelo no presenta una ralacion directa donde pueda ppredecir de manera certera los valores de colesterol considerando todas las variables, pues el R2 esta muy bajito, esto quiere decir que el colesterol total de una persona no es predecible por el comportamiento de las variables por el corazon, colesterol, genero ni cigarrilos por dia. No estaria bien utilizar estas variables para conocer el colesterol total de una persona, los valores estan realmente dispersos.

9. Construye un modelo de regresión logística para predecir el riesgo de sufrir una enfermedad coronaria en los próximos diez años, a partir de las variables descritas. Verifica el balanceo de datos y evalúa tu modelo.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

dfObsRL = dfObs.dropna(subset=["totChol", "M", "F", "cigsPerDay", "sysBP", "heartRate", "BMI", "age", "prevalentHyp", "TenYearCHD"])

#declarar las variables dependiente e independientes
x = dfObsRL[["age", "totChol", "prevalentHyp", "cigsPerDay", "sysBP", "heartRate", "BMI"]].apply(pd.to_numeric, errors="coerce")
#se pasa a datos numericos el ritmo cardiaco y en la mismo formulase convierte en valores binarios
y = pd.to_numeric(dfObsRL["TenYearCHD"])

#conjunto de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=32, stratify=y)

#crear modelo
model = LogisticRegression(solver="lbfgs", max_iter=1000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

#creacion de datos para grafica colocando a x en 10 años mas
X_range = np.linspace(x["age"].min(), (x["age"].max())+60, 200)

X_plot = pd.DataFrame({
    "age" : X_range,
    "totChol" : [x["totChol"].mean()] * len(X_range),
    "prevalentHyp" : [x["prevalentHyp"].mean()]*len(X_range),
    "cigsPerDay": [x["cigsPerDay"].mean()]*len(X_range),
    "sysBP": [x["sysBP"].mean()]*len(X_range),
    "heartRate": [x["heartRate"].mean()]*len(X_range),
    "BMI" : [x["BMI"].mean()]*len(X_range)
})
#modelo de prediccion
y_prob = model.predict_proba(X_plot)[:,1]

plt.figure(figsize=(10,3))
plt.scatter(x["age"], y, color="brown", label="Datos", s=2)
plt.plot(X_range, y_prob, color="green", linewidth=1, label="linea RLog")
plt.title("Regresion logistica para enfermedad coronaria", fontsize=12)
plt.xlabel("Edad", fontsize=10)
plt.ylabel("Probabilidad de enfermedad", fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend()
plt.show()

#Resultado
print("Exactitud (accuracy):", accuracy_score(y_test, y_pred))
print("\nMatriz de confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred))
print("\nReporte roc:\n", roc_auc_score(y_test, y_pred))
```

El modelo clasifica bien los datos en 0 o clase 0, debe ser que para el model existe la mayor cantidad de datos para los primeros 70 años pero no para 10 años mas tarde, por eso la presicion y mas confiable esta en la clase 0:
