6. Construye un gráfico para representar la población total de los países y su producto
interno bruto. Prepara los datos y escoge el gráfico adecuado (si es necesario, limpia y/o transforma los datos). Justifica.

```python
#limpieza de datos GDP
wd["GDP"] = wd["GDP"].astype(str).str.replace('[\$,]', '', regex=True).astype(float)
wd["Population"] = wd["Population"].astype(str).str.replace('[\$,]', '', regex=True).astype(float)
wd['Country'] = wd['Country'].astype(str)
wd[["Population", 'GDP']]
```

```python
# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(7, 50))

# Gráfico de barras horizontal
wd = wd.sort_values(by='Population', ascending=False)
sns.barplot(y="Country", x="Population", data=wd, ax=ax1, color="skyblue", label='POBLACION')

# Grafcio de puntos
ax2 = ax1.twiny()
sns.scatterplot(y="Country", x="GDP", data=wd, ax=ax2, color="green", markers='o', label="PIB")

# Etiquetas y título
ax1.set_xlabel("Población")
ax2.set_ylabel('PIB')
ax1.set_ylabel("País")
ax1.set_title("Población y PIB por País")
fig.legend(loc='upper left')
ax1.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

7. Escoge otro par de datos, límpialos, prepáralos y analiza su posible relación. Justifica
utilizando las herramientas vistas y gráficos adecuados.

```python
#Se toma como datos el tamaño de las fuerzas armadas y el tamaño en km2 del pais para hacer una relacion del pais con mayor capacidad
poder = df[["Country", "Armed Forces size", "Land Area(Km2)", "Abbreviation"]]

#Se eliminan caracteres de comas
poder["Armed Forces size"] = poder["Armed Forces size"].str.replace(r'[ ,]', "", regex=True)
poder["Land Area(Km2)"] = poder["Land Area(Km2)"].str.replace(r'[ ,]', "", regex=True)

#Se convierta datos  a datos numericos
poder["Armed Forces size"] = pd.to_numeric(poder["Armed Forces size"], errors="coerce")
poder["Land Area(Km2)"] = pd.to_numeric(poder["Land Area(Km2)"], errors="coerce")

#se elimina filas con celdas vacias
poder = poder.dropna()

#Se elimina las filas con campos 0 por no poderse hacer un calculo
poder = poder[(poder[["Country", "Armed Forces size", "Land Area(Km2)"]] != 0).all(axis=1)]
```

```python
poder
```

```python
#Se calcula cantidad de efectivos militares por Km2
poder["Armed Forces per Km2"] = poder["Armed Forces size"] / poder["Land Area(Km2)"]
#se ordena de Mayor a menor
poder = poder.sort_values(by=["Armed Forces per Km2"], ascending=False)
poder = poder.head(30)
poder
```

```python
#se importa tema
sns.set_theme(style="whitegrid")

#Se grafica
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 7))
#f, ax2 = plt.subplots(figsize=(7, 7))

#SE DEFINE LOS GRAFICOS

sns.set_color_codes("pastel")
sns.barplot(y="Country", x="Armed Forces per Km2", ax=ax1, data=poder, label="Mens / Km2", color="b")
sns.set_color_codes("muted")
sns.barplot(y="Country", x="Land Area(Km2)", ax=ax2, data=poder, label="Km2", color="b")

#etiquetas
ax1.legend(ncol=2, loc="lower right")
ax2.legend(ncol=2, loc="lower right")
ax2.set_yticklabels([])
ax2.set_ylabel("")
ax1.set_ylabel("Paises")
ax1.set_title("PRIMEROS 30 PAISES CON MAYOR CANTIDAD DE EFECTIVOS \nMILITARES POR METRO CUADRADO", fontsize=12, fontweight="bold", color="green")


```

```python
poder1 = poder.select_dtypes(include=['number'])
poder_corr = poder1.corr()
poder_corr
```

```python
plt.Figure(figsize=(20,10))
sns.heatmap(poder_corr, annot=True, linewidths=.5)
plt.title("MAPA DE CORRELACION")
plt.show()
```
